```python
from scipy import stats as stats
branch_a =[5200,5300,5100,5000,4950]
branch_b =[4700,4600,4800,4900,4850]
t_stat, p_val = stats.ttest_ind(branch_a, branch_b)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

## Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample Data
data = {
    'Advertising_Spend': [1000, 1500, 1200, 1700, 1300],
    'Units_Sold': [45, 60, 50, 65, 55],
    'Sales_Amount': [5000, 7000, 5600, 7500, 6100]
}
df = pd.DataFrame(data)

X = df[['Advertising_Spend', 'Units_Sold']]
y = df['Sales_Amount']

model = LinearRegression()
model.fit(X, y)

print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Sample Data
data = {
    'Advertising_Spend': [1000, 1500, 1200, 1700, 1300],
    'Sales_Amount': [5000, 7000, 5600, 7500, 6100]
}
df = pd.DataFrame(data)

# Build the Model
X = df[['Advertising_Spend']]
y = df['Sales_Amount']
model = LinearRegression()
model.fit(X, y)

# Predict values for line
df['Predicted_Sales'] = model.predict(X)

# Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Advertising_Spend', y='Sales_Amount', data=df, color='blue', label='Actual')
sns.lineplot(x='Advertising_Spend', y='Predicted_Sales', data=df, color='red', label='Regression Line')
plt.title('Simple Linear Regression: Advertising vs Sales')
plt.xlabel('Advertising Spend (KES)')
plt.ylabel('Sales Amount (KES)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Multiple regression

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


df = pd.DataFrame({
    'Advertising_Spend': [1000, 1500, 1200, 1700, 1300],
    'Units_Sold': [45, 60, 50, 65, 55],
    'Sales_Amount': [5000, 7000, 5600, 7500, 6100]
})

# Prepare data
X = df[['Advertising_Spend', 'Units_Sold']]
y = df['Sales_Amount']
model = LinearRegression()
model.fit(X, y)

# Mesh grid for surface
ad_range = np.linspace(df['Advertising_Spend'].min(), df['Advertising_Spend'].max(), 10)
unit_range = np.linspace(df['Units_Sold'].min(), df['Units_Sold'].max(), 10)
ad_mesh, unit_mesh = np.meshgrid(ad_range, unit_range)
Z = model.predict(np.c_[ad_mesh.ravel(), unit_mesh.ravel()]).reshape(ad_mesh.shape)

# 3D Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Advertising_Spend'], df['Units_Sold'], y, color='blue', label='Actual Sales')
ax.plot_surface(ad_mesh, unit_mesh, Z, color='blue', alpha=0.6)
ax.set_xlabel('Advertising Spend')
ax.set_ylabel('Units Sold')
ax.set_zlabel('Sales Amount')
ax.set_title('Multiple Linear Regression: 3D Surface')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Example: Predicting for a new data point
new_data = np.array([[1400, 58]])  # shape (1, 2)
new_df = pd.DataFrame(new_data, columns=X.columns)
prediction = model.predict(new_df)
print(prediction)
```
