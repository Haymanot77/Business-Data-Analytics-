```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Date'] = pd.to_datetime(df['Date'])

df['month_num'] = df['Date'].dt.month
monthly_sales = df.groupby(['month_num'])['Sales_Amount'].sum().reset_index()
monthly_sales = monthly_sales.sort_values("month_num")

plt.figure(figsize=(10,0))
sns.lineplot(data=monthly_sales,x='month_num',y='Sales_Amount',markers='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Example: loading your data
# df = pd.read_csv('your_data.csv')

# Convert 'Date' to datetime if needed
df['Date'] = pd.to_datetime(df['Date'])

# Extract month name or number
df['Month'] = df['Date'].dt.strftime('%b')  # e.g., 'Jan', 'Feb', etc.
# Or use month number for sorting
df['Month_Num'] = df['Date'].dt.month

monthly_sales = df.groupby(['Month_Num', 'Month'])['Sales_Amount'].sum().reset_index()

# Sort by Month_Num to keep months in correct order
monthly_sales = monthly_sales.sort_values('Month_Num')



plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_sales, x='Month', y='Sales_Amount', marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```
